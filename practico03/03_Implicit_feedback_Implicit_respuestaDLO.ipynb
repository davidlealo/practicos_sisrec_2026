```python
import pandas as pd
import numpy as np
import implicit
import scipy.sparse as sparse
```

```python
columns = ['movieid', 'title', 'release_date', 'video_release_date', \
           'IMDb_URL', 'unknown', 'Action', 'Adventure', 'Animation', \
           'Children', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', \
           'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', \
           'Thriller', 'War', 'Western']
```

```python
# Primero creamos el dataframe con los datos
df_train = pd.read_csv('u2.base',
                         sep='\t',
                         names=['userid', 'itemid', 'rating', 'timestamp'],
                         header=None)

# rating >= 3 , relevante (1) y rating menor a 3 es no relevante (0)
df_train.rating = [1 if x >=3 else 0 for x in df_train.rating ]
```

```python
# Cargamos el dataset con los items
df_items = pd.read_csv('u.item',
                        sep='|',
                        index_col=0,
                        names = columns,
                        header=None,
                        encoding='latin-1')
```

```python
# Cargamos el dataset de testing
df_test = pd.read_csv('u2.test',
                      sep='\t',
                      names=['userid', 'itemid', 'rating', 'timestamp'],
                      header=None)


# rating >= 3 es relevante (1) y rating menor a 3 es no relevante (0)
df_test.rating = [1 if x >=3 else 0 for x in df_test.rating ]


user_items_test = {}

for row in df_test.itertuples():
    if row[1] not in user_items_test:
        user_items_test[row[1]] = []

    user_items_test[row[1]].append(row[2])
```

### Métricas

```python
# Definicion de métricas (No editar)

def precision_at_k(r, k):
    assert 1 <= k <= r.size
    return (np.asarray(r)[:k]).mean()

def average_precision_at_k(r, k):
    r = np.asarray(r)
    n_rel = r.sum() # Número de items relevantes
    if n_rel == 0:
        return 0.
    vectorized_precision = np.vectorize(lambda i: precision_at_k(r, i))
    indices = np.arange(1, len(r) + 1)
    precisions = vectorized_precision(indices) # Precision@k por cada posición del ranking
    score = np.sum(precisions * r)
    return score / min(k, n_rel)

def dcg_at_k(r, k):
    r = np.asarray(r)[:k]
    if r.size:
        return np.sum(np.subtract(np.power(2, r), 1) / np.log2(np.arange(2, r.size + 2)))
    return 0.


def ndcg_at_k(r, k):
    idcg = dcg_at_k(sorted(r, reverse=True), k)

    if not idcg:
        return 0.
    return dcg_at_k(r, k) / idcg
```

### Preprocesamiento de los datos a formato sparse

```python
user_items = {}
itemset = set()

for row in df_train.itertuples():
    if row[1] not in user_items:
        user_items[row[1]] = []

    user_items[row[1]].append(row[2])
    itemset.add(row[2])

itemset = np.sort(list(itemset))

sparse_matrix = np.zeros((len(user_items), len(itemset)))

for i, items in enumerate(user_items.values()):
    sparse_matrix[i] = np.isin(itemset, items, assume_unique=True).astype(int)

matrix = sparse.csr_matrix(sparse_matrix.T)

user_item_matrix = matrix.T.tocsr()
```

```python
# Mapeo de user id a fila de la matriz sparse
user2row = {user_id: matrix_row for matrix_row, user_id in enumerate(user_items.keys())}
row2user = {matrix_row: user_id for user_id, matrix_row in user2row.items()}

# Mapeo de item id a columna de la matriz sparse
item2col = {item_id: matrix_col for matrix_col, item_id in enumerate(itemset)}
col2item = {matrix_col: item_id for item_id, matrix_col in item2col.items()}
```

```python
def evaluate_model(model, n):
  mean_ap = 0. # o MAP
  mean_ndcg = 0.
  for user_id in user_items_test.keys():
    user_row = user2row[user_id]
    rec = model.recommend(user_row, user_item_matrix[user_row], n)[0]
    rec = [col2item[col] for col in rec]
    rel_vector = np.isin(rec, user_items_test[user_id], assume_unique=True).astype(int)
    mean_ap += average_precision_at_k(rel_vector, n)
    mean_ndcg += ndcg_at_k(rel_vector, n)

  mean_ap /= len(user_items_test)
  mean_ndcg /= len(user_items_test)

  return mean_ap, mean_ndcg
```

```python
def show_recommendations(model, user, n):
  recommendations = model.recommend(userid=user, user_items=user_item_matrix[user], N=n)[0]
  return df_items.loc[recommendations]['title']
```

```python
def show_similar_movies(model, item, n=10):
  sim_items = model.similar_items(item, n)[0]
  return df_items.loc[sim_items]['title']
```

## ALS (Implicit Feedback)




**Pregunta 1 (2ptos):** Explique brevemente cómo funciona el algoritmo ALS.

**Respuesta:** Este algoritmo permite hacer filtrado colaborativo basándose en elementos históricos que no tiene que ver con el rating o una evaluación directa que el mismo usuario entregue para rankear, por ejemplo el tiempo de permanencia, click, etc.

Cómo funciona? Descompone nuestra matriz de usuarios e items que es gigante porque tenemos varios items (muchos de ellos con ceros) en menos, generando características latentes que no son interpretables directamente pero que reflejan las caracteristicas del usuario (como mencioné antes tiempo de permanencia, clicks, compartidos, si vio solamente la mitad de una película, etc)


No usé IA en esto, pero si me leí esto https://medium.com/@anurag.tech/movie-recommendation-system-using-alternating-least-squares-14c7bed98f08

```python
# Definimos y entrenamos el modelo con optimización ALS
model_als = implicit.als.AlternatingLeastSquares(factors=100, iterations=10, use_gpu=False)
model_als.fit(user_item_matrix)
```

Ejemplo de recomendación y búsqueda de items similares con los factores latentes ya entrenados:

```python
show_recommendations(model_als, user=77, n=10)
```

```python
maprec, ndcg = evaluate_model(model_als, n=10)
print('map: {}\nndcg: {}'.format(maprec, ndcg))
```

**Pregunta 2 (6ptos):** Pruebe distintos valores (al menos tres combinaciones) para los parámetros de ALS y muestre gráficos de cómo se ven afectadas las métricas recién mostradas.

```python
import matplotlib.pyplot as plt

n_eval = 10

factors_list = [10, 20, 50, 100, 150, 200, 300]
map_v_factors = []
ndcg_v_factors = []

print("Evaluando impacto de Factores...")
for f in factors_list:
    model = implicit.als.AlternatingLeastSquares(factors=f, iterations=15, use_gpu=False, random_state=42)
    model.fit(user_item_matrix, show_progress=False)
    m, n = evaluate_model(model, n=n_eval)
    map_v_factors.append(m)
    ndcg_v_factors.append(n)


iters_list = [2, 5, 10, 15, 20, 30, 50]
map_v_iters = []
ndcg_v_iters = []

print("Evaluando impacto de Iteraciones...")
for i in iters_list:
    model = implicit.als.AlternatingLeastSquares(factors=50, iterations=i, use_gpu=False, random_state=42)
    model.fit(user_item_matrix, show_progress=False)
    m, n = evaluate_model(model, n=n_eval)
    map_v_iters.append(m)
    ndcg_v_iters.append(n)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

ax1.plot(factors_list, map_v_factors, marker='o', label='MAP@10', linewidth=2)
ax1.plot(factors_list, ndcg_v_factors, marker='s', label='nDCG@10', linewidth=2)
ax1.set_title('Efecto del Número de Factores\n(Iteraciones fijas en 15)')
ax1.set_xlabel('Factores Latentes')
ax1.set_ylabel('Métrica')
ax1.grid(True, linestyle='--', alpha=0.6)
ax1.legend()

ax2.plot(iters_list, map_v_iters, marker='o', label='MAP@10', color='green', linewidth=2)
ax2.plot(iters_list, ndcg_v_iters, marker='s', label='nDCG@10', color='orange', linewidth=2)
ax2.set_title('Efecto del Número de Iteraciones\n(Factores fijos en 50)')
ax2.set_xlabel('Iteraciones de ALS')
ax2.set_ylabel('Métrica')
ax2.grid(True, linestyle='--', alpha=0.6)
ax2.legend()

plt.tight_layout()
plt.show()
```

Entre mayor el número de factores bajaron tanto MAP como nDCG

```python
show_recommendations(model_als, user=77, n=10)
```

## BPR

**Pregunta 3 (2ptos):** Explique con sus palabras la intuición del framework BPR.

**Respuesta:** BRP es un algoritmo de clasificación que tomando el feedback implícito busca hacer un correcto ranking del orden de los elementos de los ítems.



```python
# Definimos y entrenamos el modelo de implicit feedback utilizando optimizacion BPR
model_bpr = implicit.bpr.BayesianPersonalizedRanking(factors=400, iterations=40, use_gpu=False)
model_bpr.fit(user_item_matrix)
```

Ejemplo de recomendación y búsqueda de items similares con los factores latentes ya entrenados:

```python
show_recommendations(model_bpr, user=77, n=10)
```

```python
show_similar_movies(model_bpr, item=171, n=10)
```

```python
maprec, ndcg = evaluate_model(model_bpr, n=10)
print('map: {}\nndcg: {}'.format(maprec, ndcg))
```

**Pregunta 4 (6ptos):** Pruebe distintos valores (al menos tres combinaciones) para los parámetros de BPR y muestre gráficos de cómo se ven afectadas las métricas de ranking (nDCG@10 y MAP) recién mostradas.

```python
bpr_params = [
    {'factors': 100, 'lr': 0.01},
    {'factors': 200, 'lr': 0.05},
    {'factors': 400, 'lr': 0.1}
]

bpr_map = []
bpr_ndcg = []
bpr_labels = []

for p in bpr_params:
    model = implicit.bpr.BayesianPersonalizedRanking(
        factors=p['factors'],
        learning_rate=p['lr'],
        iterations=50,
        use_gpu=False,
        random_state=42
    )
    model.fit(user_item_matrix, show_progress=False)

    m, n_val = evaluate_model(model, n=10)
    bpr_map.append(m)
    bpr_ndcg.append(n_val)
    bpr_labels.append(f"F:{p['factors']}, LR:{p['lr']}")

plt.figure(figsize=(10, 6))
plt.plot(bpr_labels, bpr_map, marker='o', label='MAP@10')
plt.plot(bpr_labels, bpr_ndcg, marker='s', label='nDCG@10')
plt.title('Rendimiento BPR vs Parámetros')
plt.xlabel('Configuración (Factores y Learning Rate)')
plt.ylabel('Métrica')
plt.legend()
plt.grid(True)
plt.show()
```

Al aumentar los factores baja nuestras métricas

```python
best_model_bpr = implicit.bpr.BayesianPersonalizedRanking(
    factors=100,
    learning_rate=0.01,
    iterations=50,
    use_gpu=False,
    random_state=42
)
best_model_bpr.fit(user_item_matrix, show_progress=False)
```

```python
show_recommendations(best_model_bpr, user=77, n=10)
```
